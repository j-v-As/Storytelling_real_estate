```python
import os
import pandas as pd

# Directory containing the CSV files
csv_directory = "

# Initialize an empty DataFrame to store the merged data
merged_data = pd.DataFrame()

# Loop through files in the directory
for filename in os.listdir(csv_directory):
    if filename.startswith("gemeente-") or filename.startswith("regio-") or filename.startswith("provincie-"):
        # Load the CSV file into a DataFrame
        df = pd.read_csv(os.path.join(csv_directory, filename))
        
        # Append the data to the merged DataFrame
        merged_data = merged_data.append(df, ignore_index=True)

# Save the merged DataFrame to a new CSV file
merged_data.to_csv("merged_data.csv", index=False)
```

```python
import pandas as pd


input_file = 'merged_data.csv'


df = pd.read_csv(input_file)
df = df.drop_duplicates(subset='url')
output_file = 'Funda_final.csv'
df.to_csv(output_file, index=False)

print(f'Duplicate rows based on "url" removed. Filtered data saved to {output_file}')
```
