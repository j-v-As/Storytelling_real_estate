```python
import pandas as pd 
df = pd.read_csv("f-string-embedding.csv")  
```

```python
import re
# Function to extract the number of bedrooms
def extract_bedrooms(text):
    match = re.search(r'(\d+) slaapkamers', text)
    return int(match.group(1)) if match else None


df['num_bedrooms'] = df['num_of_rooms'].apply(extract_bedrooms)

# Function to extract the first mention of the number of rooms
def extract_rooms(text):
    match = re.search(r'(\d+) kamer(s)?', text)
    return int(match.group(1)) if match else None

df['num_rooms'] = df['num_of_rooms'].apply(extract_rooms)

# Remove rows where 'tag' is NaN
df = df.dropna(subset=['tag'])



```

```python
# Function to standardize energy labels
def standardize_energy_label(label):
    if label and isinstance(label, str):
        # Check if the label is one of the A variants with pluses
        if label.startswith('A') and '+' in label:
            return 'A'
        else:
            return label
    return None


df['energy_label_standard'] = df['energy_label'].apply(standardize_energy_label)
```

```python
### DROP ALL THE COLLUMNS#### 

# List of columns you want to drop
columns_to_drop = ['energy_label', 'price', 'address', 'descrip', 'listed_since', 'zip_code', 'size', 'year', 'living_area','kind_of_house','zip_code_4_digits','postcode','num_of_rooms', 'num_of_bathrooms', 'layout', 'insulation', 'heating', 'ownership', 'exteriors', 'parking', 'date_list', 'last_ask_price', 'last_ask_price_m2', 'city', 'log_id', 'num of tokens per descrip',]  # Replace with your actual column names

# Drop the columns
df = df.drop(columns_to_drop, axis=1)

```

```python
# Drop the columns
df = df.drop('numerical_price_per_m2', axis=1)
```

```python

df = df.drop(df.columns[0], axis=1)
```

```python
# Check for missing values and their data types
missing_info = df.isnull().sum()
missing_info = missing_info[missing_info > 0]
missing_data_types = df[missing_info.index].dtypes
missing_info, missing_data_types
```

```python
# Counting NaN values in each column
nan_counts_per_column = df.isna().sum()

# Filtering to show only columns with NaN values
columns_with_nan = nan_counts_per_column[nan_counts_per_column > 0]

columns_with_nan
```

```python

# Fill numerical columns with median
numerical_columns = ['living_area_float', 'size_float', 'latitude', 'longitude']

for col in numerical_columns:
    median_value = df[col].median()
    df[col] = df[col].fillna(median_value)

```

```python
# Fill missing values in 'num_rooms' and 'num_bedrooms' with a placeholder or median
df['num_rooms'] = df['num_rooms'].fillna(df['num_rooms'].median())
df['num_bedrooms'] = df['num_bedrooms'].fillna(df['num_bedrooms'].median())

# Convert 'num_rooms' and 'num_bedrooms' to integers
df['num_rooms'] = df['num_rooms'].astype(int)
df['num_bedrooms'] = df['num_bedrooms'].astype(int)

# Check the data types to confirm the conversion
df.dtypes
```

```python
df_encoded = pd.get_dummies(df, columns=['building_type', 'tag', 'house_category'])


# Ordinal encoding for 'energy_label_standard'
# Defining the order of energy labels
energy_label_order = {'G': 0, 'F': 1, 'E': 2, 'D': 3, 'C': 4, 'B': 5, 'A': 6, 'na': -1,'null':-1,"NaN":-1}
df_encoded['energy_label_encoded'] = df['energy_label_standard'].map(energy_label_order)

# Display the first few rows of the encoded dataset
df_encoded.head()
```

```python
df_encoded.drop('energy_label_standard',axis= 1)
df_encoded.drop('descrip_en',axis= 1)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler


# Initialize scalers
scaler_num = StandardScaler()
scaler_long = StandardScaler()
scaler_lat = StandardScaler()

# Scale numerical variables
df_encoded['size_scaled'] = scaler_num.fit_transform(df_encoded[['size_float']])


# Scale longitude and latitude
df_encoded['longitude_scaled'] = scaler_long.fit_transform(df_encoded[['longitude']])
df_encoded['latitude_scaled'] = scaler_lat.fit_transform(df_encoded[['latitude']])

# Drop columns

columns_to_drop = [ 'size_float', 'living_area_float', 'latitude','longitude','energy_label_standard']
df_encoded.drop(columns=columns_to_drop, inplace=True)
```
